```python
import numpy as np 
import pandas as pd
import re
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

df = pd.read_csv('you_tube_comment_1.csv')

df.head()
```

```python
custom_stopwords = set(STOPWORDS)
indonesian_stopwords = {'ini', 'itu', 'saya', 'dia', 'kamu', 'kita', 'mereka', 'aku', 'nya', 'dan', 'yang', 'untuk', 'dengan', 'atau', 'dalam', 'dari', 'kan', 'dulu', 'tahun', 'di' }
custom_stopwords.update(indonesian_stopwords)
```

```python
# Menggabungkan semua komentar menjadi satu string
all_comments = " ".join(df['comment'].astype(str))
```

```python

# Membersihkan teks
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Menghilangkan karakter spesial
    text = re.sub(r'\d+', '', text)  # Menghilangkan angka
    text = re.sub(r'ðŸ[^\s]*', '', text)  # Menghilangkan karakter unicode yang rusak
    return text

cleaned_comments = clean_text(all_comments)
```

```python
cleaned_comments
```

```python
wordcloud = WordCloud(width=1000, height=500, background_color='white', colormap='viridis',  stopwords=custom_stopwords).generate(cleaned_comments)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
